```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_animated_films_in_the_United_States_and_Canada'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[5]
```

```python
table = soup.find_all('table')[5]
```

```python
print(table)
```

```python
anime_titles = table.find_all('th')
```

```python
anime_titles
```

```python
anime_table_titles = [title.text.strip() for title in anime_titles]

print(anime_table_titles)
```

# Using Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = anime_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
